```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./wine_data.csv", header=None, usecols=[0,1,2])
df.columns=['Brand', 'Alcohol', 'Malic Acid']
```

```python
sns.kdeplot(data=df['Alcohol'])
sns.kdeplot(data=df['Malic Acid'])
```

```python
sns.scatterplot(data=df, x='Alcohol', y='Malic Acid', hue='Brand')
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(df.drop('Brand', axis=1), df['Brand'], train_size=0.8)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(xTrain)

xTrainScaled = scaler.transform(xTrain)
xTestScaled = scaler.transform(xTest)
```

```python
xTrainScaled=pd.DataFrame(xTrainScaled, columns=xTrain.columns)
xTestScaled=pd.DataFrame(xTestScaled, columns=xTest.columns)
```
