```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('house-train.csv',usecols=['GarageQual','FireplaceQu', 'SalePrice'])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in data['GarageQual'].unique():
    sns.kdeplot(data[data['GarageQual']==i]['SalePrice'],label=i)

plt.legend()
plt.show()
```

Now lets fill the null falues with randomly generated data

```python
data['GarageQualImputed'] = data['GarageQual']
```

```python
data['GarageQualImputed'][data['GarageQualImputed'].isnull()] = data['GarageQualImputed'].dropna().sample(data['GarageQualImputed'].isnull().sum()).values
```

```python
data.isnull().sum()
```

Here, we cans see that there are no missing values in the GarageQualImputed column

```python
for i in data['GarageQual'].unique():
    sns.kdeplot(data[data['GarageQual']==i]['SalePrice'],label=i)
    sns.kdeplot(data[data['GarageQualImputed']==i]['SalePrice'],label=str(i)+"Imputed")

plt.legend()
plt.show()
```

We can see that there is very minimal difference between the distribution before imputation and after imputation

similarly, we can randomly impute values in the FirePlaceQu column

```python
data['FireplaceQuImp'] = data['FireplaceQu']
```

```python
data.head()
```

```python
data['FireplaceQuImp'][data['FireplaceQuImp'].isnull()] = data['FireplaceQuImp'].dropna().sample(data['FireplaceQuImp'].isnull().sum()).values
```

```python
data.isnull().sum()
```

Now we can see that there are no missing values in the imputed columns

```python
for i in data['FireplaceQu'].unique():
    sns.kdeplot(data['FireplaceQu']==i, label = i)
    sns.kdeplot(data['FireplaceQuImp']==i, label = str(i)+"Imputed")

plt.legend()
plt.show()
```
