```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("./train.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

```python
df.head()
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(df.drop(columns=['Survived']), df['Survived'], train_size=0.8, random_state=2)
```

```python
xTrain
```

```python
numColumns = ['Age','Fare']
catColumns = ['Sex', 'Embarked']
```

```python
pipe1 = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]
)

pipe2 = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder())
    ]
)

trf = ColumnTransformer(
    [
        ('t1', pipe1, numColumns),
        ('t2', pipe2, catColumns)
    ],
    remainder='passthrough'
)
```

```python
xTrainTransformed = trf.fit_transform(xTrain)
xTestTransformed = trf.transform(xTest)
```

```python
clf = LogisticRegression()
```

```python
clf.fit(xTrainTransformed, yTrain)
```

```python
yPred = clf.predict(xTestTransformed)
```

```python
accuracy_score(yPred, yTest)
```
