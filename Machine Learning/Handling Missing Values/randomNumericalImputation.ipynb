```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./train.csv", usecols=['Age', 'Fare', 'Survived'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.kdeplot(df['Age'])
```

Now lets make a new column and impute new values using Random Simple Imputation

```python
df["Age Imputed"]=df['Age']
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Age'].dropna().sample(df['Age Imputed'].isnull().sum())
```

```python
df['Age Imputed'][df['Age Imputed'].isnull()] = df['Age'].dropna().sample(df['Age Imputed'].isnull().sum()).values
```

```python
df.isnull().sum()
```

```python
sns.kdeplot(df['Age'], label = 'Original')
sns.kdeplot(df['Age Imputed'], label = 'Imputed')

plt.legend()
plt.show()
```

we can see a minimilistic difference between the original distribution and the new distribution of the data set

## Now lets check if some outliers were created or not

```python
sns.boxplot(df[['Age', 'Age Imputed']])
```

Also, here we can see that there is not so big differenct between the original distribution and new distribution of the data set
