```python
tips = sns.load_dataset('tips')
```

```python
iris = sns.load_dataset('iris')
```

```python
titanic = pd.read_csv('./train.csv')
```

# Scatterplot (Numerical - Numerical)

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

# Bar plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(data=titanic, x='Pclass', y='Age', hue=titanic['Sex'])
```

# Box Plot (Numerical, Categorical)

```python
sns.boxplot(data=titanic, x='Age', y='Sex', hue=titanic['Survived'])
```

# KDE Plot (Numerical - Categorical)

```python
sns.kdeplot(data=titanic[titanic['Survived']==1], x='Age', label = 'Survived')
sns.kdeplot(data=titanic[titanic['Survived']==0], x='Age', label = 'Dead')
plt.legend()
plt.show()
```

# Heatmap (Categorical Categorical)

```python
heatmapData = pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(heatmapData)
```

## We can also do the above like the followinf with more accuracy on the data with the help of percentage of people dead or alive in a particular class

```python
aliveVSclassData = titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
plt.pie(aliveVSclassData, autopct='%.2f')
```

```python
plt.bar(aliveVSclassData, titanic['Pclass'])
```

```python
aliveVSclassData.plot(kind='bar')
```
