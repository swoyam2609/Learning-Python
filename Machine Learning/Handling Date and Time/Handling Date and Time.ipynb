## Converting to date time column

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

Extracting year from the column

```python
date['year'] = date['date'].dt.year
```

```python
date.head()
```

## Extract month from the column

```python
date['Month'] = date['date'].dt.month
date['Month Name'] = date['date'].dt.month_name()
```

```python
date.head(10)
```

## Extacting the date

```python
date['exact date'] = date['date'].dt.day
```

```python
date.head()
```

## Extracting the day name

```python
date['day'] = date['date'].dt.day_name()
```

```python
date.head()
```

## Checking if its weekend or not

```python
date['weekend'] = np.where(date['day'].isin(['Sunday', 'Saturday']), 1, 0)
```

```python
date.head()
```

## Extract the week of the year

```python
date['week'] = date['date'].dt.week
```

```python
date.head()
```

## Extract the quarter

```python
date['quarter'] = date['date'].dt.quarter
```

```python
date.head()
```

## Extracting the semester

```python
date['semester'] = np.where(date['quarter'].isin([2,3]),1,2)
```

```python
date.head()
```
