```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
df['number']=df['number'].replace('A', '0')
# we replaced the characted 'A' with '0'
```

```python
df['number'] = df['number'].astype(int)
# now we converted all the characters to its respective integers
```

```python
df.head()
```

```python
df['number'].unique()
```

## Till now, we have fixed the 'number' column of the data set
now lets fix the 'cabin' column of the data set

```python
df['Cabin Class']=df['Cabin'].str[0]
df['Cabin Number']=df['Cabin'].str.extract('(\d+)')
df = df.drop(columns=['Cabin'])
```

```python
df.head()
```

```python
df['Cabin Class'].value_counts().plot(kind='bar')
```

```python
df['Ticket Type']=df['Ticket'].apply(lambda s : s.split(" ")[0])
df['Ticket Number']=df['Ticket'].apply(lambda s : s.split(" ")[-1])
```

```python
df.head()
```

```python
df['Ticket Number'] = pd.to_numeric(df['Ticket Number'], errors='coerce', downcast='integer')
```

```python
df.head(20)
```

```python
df['Ticket Type'] = np.where(df['Ticket Type'].str.isdigit(), np.NaN, df['Ticket Type'])
```
